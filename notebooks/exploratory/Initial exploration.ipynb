```python
def print_proportion_y(y):
    unique, counts = np.unique(y, return_counts=True)
    length = y.shape[0]
    d = dict(zip(unique, counts))
    print(f'{100 * (d[0] / length):.2f} % positive values')
    print(f'{100 * (d[1] / length):.2f} % negative values')

print_proportion_y(ds.y_train)
```
